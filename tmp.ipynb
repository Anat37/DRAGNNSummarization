```python
import json
import codecs

name_list = []
ADM3_NAME_list = []
with codecs.open("Regions.txt", "r", encoding='utf8') as read_file:
    data = json.load(read_file)
    for f in data['features']:
        name_list.append(f["properties"]["NAME"])
        ADM3_NAME_list.append(f["properties"]['ADM3_NAME'])
```

```python
with open('regions.txt', 'w') as f:
    f.write(name_list.__str__())
    
with open('okrug.txt', 'w') as f:
    f.write(ADM3_NAME_list.__str__())
```
